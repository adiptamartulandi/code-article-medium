```python
import pandas as pd
import warnings

from elasticsearch import Elasticsearch
from elasticsearch import helpers

warnings.filterwarnings('ignore')
```

```python
# Connect ke server Elasticsearch di localhost port 9200
es = Elasticsearch([{'host': 'localhost', 'port': 9200}])
es.info()
```

### Basic query menggunakan match dan term

```python
#get index mappings
es.indices.get(index='flipkart_products')
```

```python
#query match field product_name
es.search(
    index='flipkart_products',
    body={
        "query": {
            "match": {
                "product_name": "mobile"
            }
        }
    }
)
```

```python
#query term field brand.keyword
es.search(
    index='flipkart_products',
    body={
        "query": {
            "term": {
                "brand.keyword": "Amaze Mobile"
            }
        }
    }
)
```

### Sort dan Pagination

```python
#query with sort based on retail_price
es.search(
    index='flipkart_products',
    body={
        "query": {
            "match": {
                "product_name": "mobile"
            }
        },
        "sort": [
            {
                "retail_price": {
                    "order": "desc"
                }
            }
        ]
    }
)
```

```python
#query with sort based on retail_price and have pagination
es.search(
    index='flipkart_products',
    body={
        "query": {
            "match": {
                "product_name": "mobile"
            }
        },
        "sort": [
            {
                "retail_price": {
                    "order": "desc"
                }
            }
        ],
        "from": 0,  # Mulai dari hasil pertama
        "size": 5  # Kembalikan 10 hasil
    }
)
```

### Fleksibel query menggunakan boolean query

#### boolean query must

```python
#query with pagination and boolen query must
es.search(
    index='flipkart_products',
    body={
        "query": {
            "bool": {
                "must": {
                    "match": {
                        "product_name": "mobile phone"
                    }
                }
            }
        },
        "from": 0,  # Mulai dari hasil pertama
        "size": 5  # Kembalikan 10 hasil
    }
)
```

#### boolean query should

```python
#query with pagination and boolen query should
es.search(
    index='flipkart_products',
    body={
        "query": {
            "bool": {
                "should": [
                    {
                        "match": {
                            "product_name": "mobile phone"
                        }
                    },
                    {
                        "match": {
                            "brand.keyword": "Samsung"
                        }
                    }
                ]
            }
        },
        "from": 0,  # Mulai dari hasil pertama
        "size": 5  # Kembalikan 10 hasil
    }
)
```

#### query boolean must not

```python
#query with pagination and boolen query must_not
es.search(
    index='flipkart_products',
    body={
        "query": {
            "bool": {
                "must_not": {
                    "match": {
                        "product_name": "Multicolor Paper Sky"
                    }
                }
            }
        },
        "from": 0,  # Mulai dari hasil pertama
        "size": 5  # Kembalikan 10 hasil
    }
)
```

#### query boolean filter

```python
#query with pagination and boolen query filter
es.search(
    index='flipkart_products',
    body={
        "query": {
            "bool": {
                "filter": [
                    {
                        "range": {
                            "retail_price": {
                                "gte": 100000
                            }
                        }
                    },
                    {
                        "term": {
                            "product_name": "durian"
                        }
                    }
                ]
            }
        },
        "from": 0,  # Mulai dari hasil pertama
        "size": 5  # Kembalikan 5 hasil
    }
)
```

#### kombinasi boolean query

```python
#query with pagination and boolen query match + filter
es.search(
    index='flipkart_products',
    body={
        "query": {
            "bool": {
                "must": {
                    "match": {
                        "product_name": "mobile phone"
                    }
                },
                "filter": {
                    "range": {
                        "retail_price": {
                            "gte": 500
                        }
                    }
                }
            }
        },
        "from": 0,  # Mulai dari hasil pertama
        "size": 5  # Kembalikan 5 hasil
    }
)
```
